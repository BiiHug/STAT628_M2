```python
Text=dfAZReviewTrain['text']
len(Text)
```

```python
Text=Text[:500]
```

```python
#去除标点符号\n等,小写
Text1=[]
for i in Text:
    Text1.append(re.sub(r"[0-9\s+\.\!\/_,$%^*()?;；:+\"\']+|[+——！，;:。？、~@#￥%……&*（）]+", " ", i).lower())
```

```python
Text1[0]
```

```python
#分词
Text2=[]
for i in Text1:
    Text2.append(i.split())
```

```python
#lemmatization词性还原
import nltk.stem as ns
lemmatizer = ns.WordNetLemmatizer()

Text4=[]
for text in Text2:
    # 将名词还原为单数形式
    Text3=[]
    for i in text:
        n_lemma = lemmatizer.lemmatize(i, pos='n')
        Text3.append(n_lemma)
    Text4.append(Text3)
```

```python
Text6=[]
for text in Text4:
    # 将动词还原为原型形式
    Text5=[]
    for i in text:
        v_lemma = lemmatizer.lemmatize(i, pos='v')
        Text5.append(v_lemma)
    Text6.append(Text5)
```

```python
##stemming词干还原
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
porter_stemmer= PorterStemmer()
Text8=[]
for text in Text6:
    Text7=[]
    for i in text:
        Text7.append(porter_stemmer.stem(i))  
    Text8.append(Text7)
```

```python
#去除停词
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

Text10=[]
for text in Text8:
    Text9 = [i for i in text if not i in stop_words]
    Text10.append(Text9)
```

```python
def connect(text):
    for x in text:
        return " ".join(text)
text_for_analysis = []
for i in range(len(Text10)):
    text_for_analysis.append(connect(Text10[i]))
```
